```python
import optuna
import optuna.visualization as vis


def objective(trial):
    x = trial.suggest_float('x', -10, 10)
    y = trial.suggest_float('y', -10, 10)
    return (x - y) ** 2

study = optuna.create_study()
study.optimize(objective, n_trials=100)
```

```python


print(study.best_params)  # E.g. {'x': 2.002108042}

vis.plot_param_importances(study)   # which params matter most
vis.plot_contour(study)             # where in the space it's focusing
vis.plot_optimization_history(study) # is it actually improving?
```
